#### **Bar char**


````python
values = [10,24,15,20]
categories = ['red', 'yellow', 'green', 'blue']

plt.barh(categories, values, color=categories, height=0.9)
plt.title('Bar Char Title')
plt.ylabel('Categoires')
plt.xlabel('Values')
plt.show()
````

__Double bar char__

````python
categories = ['A', 'B', 'C', 'D']
values1 = [13,24,17,22]
values2 = [27,15,19,12]

x = np.arange(len(categories))

plt.bar(x-0.16, values1, color='b', width=0.3, label='Label 1')
plt.bar(x+0.16, values2, color='r', width=0.3, label='Label 2')

plt.title('Bar Char Title')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()

plt.show()
````

__Stacked Barchar__

````python
x = np.arange(len(categories))

plt.bar(x, values1, color='b', width=0.8, label='Label 1')
plt.bar(x, values2, color='r', bottom=values1, width=0.8, label='Label 2')

plt.title('Bar Char Title')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()

plt.show()
````

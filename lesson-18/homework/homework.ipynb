From movie.csv:

Identify the director with the highest total director_facebook_likes.
Find the 5 longest movies and their respective directors.

```python
director_likes = df_movies.groupby('director_name')['director_facebook_likes'].sum()
top_director = director_likes.idxmax()
print(f"Top liked director is '{top_director}' his likes:", director_likes.max())
sorted_movies = df_movies.sort_values(by=['director_facebook_likes'])
sorted_movies
```
